```python
import pygame
```

```python
pygame.init()
gameDisplay = pygame.display.set_mode((800,600))
pygame.display.set_caption('Bit Race')
clock = pygame.time.Clock()
```

```python
crashed = False

while not crashed:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            crashed = True
            
        print (event)
        
    pygame.display.update()    
    clock.tick(60)
    
pygame.quit()
    
```
